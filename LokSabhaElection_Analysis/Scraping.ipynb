### State-wise Parlimentary constituencies and their winners

```python
url = 'https://results.eci.gov.in/PcResultGenJune2024/index.htm'
res = requests.get(url)
soup = bs(res.content,'html.parser')
```

```python
state_code = []
selects = soup.find('select',{'id':"ctl00_ContentPlaceHolder1_Result1_ddlState"})
options = selects.find_all('option')
for option in options:
    state_value = option.get('value')
    state_link = 'https://results.eci.gov.in/PcResultGenJune2024/partywiseresult-'+state_value+'.htm'
    state_name = option.text.strip()
    if state_value:  # Skip the first option as it is just a placeholder
        state_code.append((state_link, state_name))
print(state_code)
```

```python
statewise_list = []
def getWinners(link,name):
    win_candidate_list = []
    party_res = requests.get(link)
    party_soup = bs(party_res.content,'html.parser')
    win_candidate_table = party_soup.find('table')
    for row in win_candidate_table.findAll('tr')[1:]:
        cells = row.findAll('td')[1:]
        row_data = [cell.text.strip() for cell in cells]
        row_data.insert(0,name)
        win_candidate_list.append(row_data)
    return win_candidate_list
for link,name in state_code:
    res = requests.get(link)
    soup = bs(res.content,'html.parser')
    table = soup.find('table')
    anchor_tags = table.find_all('a')
    links = []
    party_wise_links = []
    for tag in anchor_tags:
        link = tag.get('href')
        if link:
            links.append(link)
    party_wise_links = ['https://results.eci.gov.in/PcResultGenJune2024/'+link for link in links]
    for i in party_wise_links:
        winners = getWinners(i,name)
        statewise_list.append(winners)
print(statewise_list)
```

```python
flattened_data = [item for sublist in statewise_list for item in sublist]
print(flattened_data)
```

```python
columns = ['StateName', 'Parliament Constituency', 'Winning Candidate', 'Total Votes', 'Margin']
state_con_list = pd.DataFrame(flattened_data, columns=columns)
state_con_list.head()
```

```python
winnersList = pd.merge(state_con_list,winner_list,on=['Parliament Constituency', 'Winning Candidate', 'Total Votes', 'Margin'])
winnersList.head()
```

```python
winnersList.to_csv('winners_list.csv',index=False)
```
